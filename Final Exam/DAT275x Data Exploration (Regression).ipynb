```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
pd.set_option('max_rows',20000)
pd.set_option('max_columns',30)
```

```python
df = pd.read_csv("regression.csv")
```

### Plotting graphs

```python
plt.figure(figsize=(15,12))
sns.boxplot(x=df.YearlyIncome,data=df)
plt.show()
```

```python
plt.figure(figsize=(15,12))
sns.boxplot(df.Age,data=df)
plt.show()
```

```python
plt.figure(figsize=(15,12))
sns.boxplot(df.AveMonthSpend,data=df)
plt.show()
```

```python
sns.pairplot(df.sample(1000))
plt.show()
```

```python
plt.figure(figsize=(15,12))
sns.relplot(x='Age',y='AveMonthSpend',data=df,kind='scatter',hue='Gender',palette='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(15,12))
sns.relplot(x='YearlyIncome',y='AveMonthSpend',data=df,kind='scatter',hue='Gender',palette='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(25,15))
sns.lmplot(x='YearlyIncome',y='AveMonthSpend', data=df, palette='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(15,12))
sns.jointplot(x=df.YearlyIncome,y=df.AveMonthSpend,data=df, kind="reg")
plt.show()
```

```python
plt.figure(figsize=(15,12))
sns.jointplot(x=df.Age,y=df.AveMonthSpend,data=df, kind="reg")
plt.show()
```

```python
plt.figure(figsize=(15,12))
sns.countplot(x=df.Education,data=df)
plt.show()
```

```python
plt.figure(figsize=(15,12))
sns.countplot(x=df.Education,data=df, hue='Occupation')
plt.show()
```

```python
plt.figure(figsize=(15,12))
sns.stripplot(x=df.Education,y=df.AveMonthSpend,data=df, jitter=False, hue=df.Gender)
plt.show()
```
